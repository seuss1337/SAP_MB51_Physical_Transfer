```
import pandas as pd
import numpy as np
import openpyxl as xl
import os
import shutil
import sys
print("Property of JP Enciso-Siller Employee 30428 of 3900")
from os import listdir
from os.path import isfile, join
from openpyxl.styles.borders import Border, Side
```

```
#Get list of files in Input
files = [f for f in listdir(r'/content/drive/My Drive/Colab Notebooks/MB51/Input/') if isfile(join(r'/content/drive/My Drive/Colab Notebooks/MB51/Input/', f))]
source_dir ='/content/drive/My Drive/Colab Notebooks/MB51/Resource/MASTER.xlsx'
movements_df = pd.read_csv (r'/content/drive/My Drive/Colab Notebooks/MB51/Resource/job_names.csv')

tborder = Border(left=Side(style='thin'),
                     right=Side(style='thin'),
                     top=Side(style='thin'),
                     bottom=Side(style='thin'))


#loop through every file in /Input  ###MAIN LOOP
for file in files:
  print("Working on "+join(r'/content/drive/My Drive/Colab Notebooks/MB51/Input/', file))
  destination_dir= join(r'/content/drive/My Drive/Colab Notebooks/MB51/Output/', file)
  movements1_df = pd.read_excel (join(r'/content/drive/My Drive/Colab Notebooks/MB51/Input/', file))
  doc_nums_raw = movements1_df["Material Document"].tolist()
  doc_nums = [] #list of doc nums from mb51
  
  shutil.copy(source_dir,destination_dir)
  wb = xl.load_workbook(destination_dir) 


  #make a list of transfers and consumptions only
  for i in doc_nums_raw:
    if str(i)[:3] == "490" and (i in doc_nums) == False:
      doc_nums.append(np.int64(i))
  print(doc_nums)

  #Loop through all document numbers 
  for x in range(len(doc_nums)):
    current_df = movements1_df.loc[movements1_df["Material Document"]  == doc_nums[x]]

    in_sheet = False
    ws = wb.worksheets[0]
    wb.copy_worksheet(ws)
    wb.worksheets[x+1].title = str(doc_nums[x])
    ws =  wb.worksheets[x+1]
    storage_location = current_df.iloc[0]['Storage Location']
    transfer_date = current_df.iloc[0]['Posting Date']
    job_name = movements_df.loc[movements_df["Storage Location"]  == storage_location].iloc[0]['Job Name']
    job_name =job_name[:44]
    job_number = movements_df.loc[movements_df["Storage Location"]  == storage_location].iloc[0]['Job Number']
    
    #Row possition for items not in list
    not_in_list_counter = 55

    #Fill header information
    ws.cell(row = 2, column = 2, value = job_name) #Job Name
    ws.cell(row = 2, column = 5, value = job_number) #Job Number
    ws.cell(row = 3, column = 5, value = transfer_date) #DATE
    ws.cell(row = 4, column = 5, value = str(doc_nums[x])) #Doc Number
  
    # check if material is transfer in or out 
    if int(current_df['Movement Type'].iloc[0]) == 221:
       ws.cell(row = 8, column = 6, value = "x")
    elif int(current_df['Qty in Un. of Entry'].iloc[0]) < 0:
      ws.cell(row = 7, column = 6, value = "x")
    else: ws.cell(row = 6, column = 3, value = "x")
    
    for index, row in current_df.iterrows():
      #loop through destination file and write material quantities
      
      
      for i in range(12, 65):
        
        if str(ws.cell(row = i, column = 1).value) == str(row['Material']):
          ws.cell(row = i, column = 3, value = row['Qty in Un. of Entry'])
          ws.cell(row = i, column = 3).border = tborder
          in_sheet = True
          
        if str(ws.cell(row = i, column = 4).value) == str(row['Material']):
          ws.cell(row = i, column = 6, value = row['Qty in Un. of Entry'])
          ws.cell(row = i, column = 6).border = tborder
          in_sheet = True
          
      if in_sheet == False:
        ws.cell(row = not_in_list_counter, column = 4, value = row['Material'])
        ws.cell(row = not_in_list_counter, column = 4).border = tborder
        ws.cell(row = not_in_list_counter, column = 6, value = row['Qty in Un. of Entry'])
        ws.cell(row = not_in_list_counter, column = 6).border = tborder
        not_in_list_counter += 1
      in_sheet = False
          
      

    ws.print_area = "A1:F65"
  del wb['POLISH']
  wb.save(str(destination_dir))
  print("Done Writing File: "+ file)

print("Script Complete")

```
